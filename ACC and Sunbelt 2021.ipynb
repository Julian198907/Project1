```python
#to get texas-state 2021 schedule
url = ("https://www.sports-reference.com/cfb/schools/texas-state/2021-schedule.html")
d1 = pd.read_html(url)
len(d1)
```

```python
TXst = d1[0]
```

```python
TXst = TXst.drop(columns = ['Time', 'Day', 'School', 'W', 'L', 'Streak', 'Notes'])
TXst
```

```python
#l = list(TXst.columns) to list all columns
TXst.rename(columns = {'Unnamed: 5': 'HomeAway', 'Unnamed: 8': 'WinLoss'}, inplace=True)
```

```python
TXst.fillna('H', inplace=True)
```

```python
TXst['Home Team'] = 'Texas St'
TXst.set_index('Home Team', inplace=True)
TXst.reset_index(inplace=True)
TXst.loc[(TXst.HomeAway == '@') | (TXst.WinLoss == 'W'), 'Rank'] = 1
```

```python
TXst.fillna(0, inplace=True)
TXst['weight'] = TXst['Rank'] + (TXst['Pts'] - TXst['Opp'])
TXst.head()
```

```python
#to aggregate Texas St's performace compared to other conferences
TXST1 = TXst
points = TXST1.loc[:, ('Conf','weight')]
points.sort_values(by = ['weight'])
TexasSt = points.groupby('Conf').weight.sum()
TexasSt
```

```python
#convert series to dataframe
x = pd.DataFrame(TexasSt)
```

```python
#to add column school
x.reset_index(inplace=True)
TexasState = x
TexasState.drop(columns = ['index'], inplace=True)
TexasState['School'] = 'Texas St.'
TexasState.set_index('School', inplace=True)
TexasState.reset_index(inplace=True)
TexasState
```
